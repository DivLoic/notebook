```python
import pandas as pd
import seaborn as sns

from bokeh.io import push_notebook, output_notebook, show
from bokeh.plotting import figure
output_notebook()
```

```python
FILE_PATH_TAG = u'data/kudu_tag.csv'
FILE_PATH_MOWER = u'data/kudu_mower.csv'
```

```python
### Web Tag Example
```

```python
df = pd.read_csv(FILE_PATH_TAG, sep=';')
```

```python
df.dtypes
```

```python
result = df.groupby(["browser", "is_mobile"]).count()
```

```python
result.reset_index(level=0, inplace=True)
result.reset_index(level=0, inplace=True)
result
```

```python
sns.barplot(x="browser", y="datetime",  hue="is_mobile", data=result)
```
