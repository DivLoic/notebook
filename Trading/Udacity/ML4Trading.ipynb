```python
# Rooling statistics
```

```python
AAPL = pd.read_csv("data/AAPL.csv", index_col='Date')[["Close"]]
AAPL.sort_index(ascending=True, inplace=True)
```

```python
AAPL["r_mean"] = pd.stats.moments.rolling_mean(AAPL["Close"], window=20)
AAPL["upper_std"] = AAPL["r_mean"] + 2*pd.stats.moments.rolling_std(AAPL["Close"], window=20)
AAPL["lower_std"] = AAPL["r_mean"] - 2*pd.stats.moments.rolling_std(AAPL["Close"], window=20)
```
