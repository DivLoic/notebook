#Ops

```python
target = []
c = codecs.open(File, 'r', encoding='utf-8')
#with open(File) as csv:
for line in c:
    row = line.split(";")
    array = {}
    array["last_name"] = row[0]
    array["first_name"] = row[1]
    array["group"] = row[3][0:3]
    array["pseudo"] = (row[1][0] + row[0][:7]).lower()
    array["email"] = (row[1][0] + row[0]).lower() + "@isep.fr"
    array["isep_no"] = row[2]

    target.append(array)
```

```python
df = pd.DataFrame(target)
df = df[["last_name", "first_name", "group", "pseudo", "isep_no", "email"]]
```
