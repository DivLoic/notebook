```python
import os
import sys
import pandas as pd
import random
```

```python
fileList = "/Users/LoicMDIVAD/Desktop/demo/CURRENTGROUPLIST.csv"
fakeFileList = "/Users/LoicMDIVAD/Desktop/demo/FAKEGROUPLIST.csv"
PURPLE = ["Zovena", "Delphine", "Narisely", "Loïc", "Shreya", "Matthieu"]
```

```python
fnames = ["Julien", "Justin", "Louis", "Maxime", "Michaël", "Clara", "Coralie", "Daphnée",
          "Laurence", "Laurie", "Léa", "Léanne", "Maélie", "Maéva", "Marianne", "Brandone",
         "Alban", "Alexandra","Céline"]
```

```python
df1 = pd.read_csv(fileList, sep=";")
df2 = pd.read_csv(fakeFileList, sep=";")
df3 = pd.read_csv(fakeFileList, sep=";")
```

```python
#help(pd.read_csv)
```

```python
no_col = filter(lambda y: y.find("Unnamed") != -1, df1.columns.tolist())
```

```python
df1 = df1.drop(no_col, axis=1)
df2 = df2.drop(no_col, axis=1)
df3 = df3.drop(no_col, axis=1)
```

```python
for idx, student in df1.iterrows():
    if student["prenom"] in PURPLE:
        continue
    elif student["prenom"] == df2.iloc[idx]["prenom"]:
        j = random.randint(0, len(fnames) -1)
        df3.iloc[idx, 1] = fnames[j]
        print(str(j)+ "  :  " +df2.iloc[idx]["prenom"]+ "  --->  " + fnames[j])
```

```python
for idx, student in df1.iterrows():
    if student["prenom"] in PURPLE:
        continue
    else:
        j = random.randint(0, len(lnames) -1)
        df3.iloc[idx, 0] = lnames[j]
        print(str(j)+ "  :  " +df2.iloc[idx]["nom"]+ "  --->  " + lnames[j])
```
