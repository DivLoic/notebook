```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.set(style="white", color_codes=True)
```

```python
df_iris = pd.read_csv("iris.csv")
df_iris.head()
```

```python
df_iris["species"].value_counts()
```

```python
#df_iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
sns.jointplot(x="sepal_length", y="sepal_width", data=df_iris, size=6)
```

```python
sns.FacetGrid(df_iris, hue="species", size=5) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
```

```python
sns.boxplot(x="species", y="sepal_length", data=df_iris)
sns.stripplot(x="species", y="sepal_length", data=df_iris, jitter=True, edgecolor="gray")
```

```python
sns.FacetGrid(df_iris, hue="species", size=6) \
   .map(sns.kdeplot, "sepal_length") \
   .add_legend()
```

```python
sns.pairplot(df_iris, hue="species", size=3)
```

```python
df_iris.boxplot(by="species", figsize=(24, 12))
```
